```
import trimesh
import numpy as np
from scipy.spatial import distance
import igl
import matplotlib.pyplot as plt
```

```
curved_mesh = trimesh.load('/content/drive/My Drive/draping_workspace/3d_shape.obj')
```

```
curved_mesh.show()
```

```
v = curved_mesh.vertices
f = curved_mesh.faces

p0 = 0
distances = distance.cdist(v, [v[p0]], 'euclidean')
p1 = distances.argmax()
dist = distances[p1][0]

b = np.array([2, 1], dtype='int64')

b[0] = p0
b[1] = p1

bc = np.array([[0.0, 0.0], [dist, 0.0]])

# LSCM parametrization
_, uv = igl.lscm(v, f, b, bc)
```

```
p0,p1
```

```
flat_vertices = np.vstack((uv[:,0], uv[:,1], np.zeros_like(uv[:,1]))).T
```

```
flat_mesh = trimesh.Trimesh(vertices=flat_vertices,
                       faces=curved_mesh.faces,
                       process=False)
```

```
scale_factor = curved_mesh.area/flat_mesh.area
```

```
flat_mesh.vertices = flat_mesh.vertices * np.sqrt(scale_factor) #scale like freecad
```

```
trimesh.repair.fix_normals(flat_mesh)
```

```
flat_mesh.show()
```

```
flat_mesh.export('/content/drive/My Drive/draping_workspace/2d_shape.obj')
```
